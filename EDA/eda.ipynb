# Exploratory Data Analysis (EDA)
Let's check the distribution and statistics of the target and key features in the training data.

```python
# Load training data and show basic statistics
import pandas as pd
train_df = pd.read_csv('../data/train.csv')
print('Shape:', train_df.shape)
print('Columns:', train_df.columns.tolist())
print(train_df[['units_sold']].describe())
print(train_df[['units_sold']].value_counts().head(10))
```

```python
# Show statistics for key features
key_features = ['total_price', 'base_price', 'week', 'store_id', 'sku_id']
for col in key_features:
    if col in train_df.columns:
        print(f'--- {col} ---')
        print(train_df[col].describe())
        print(train_df[col].value_counts().head(10))
```
